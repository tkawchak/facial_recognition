```python
import keras
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing.image import array_to_img, img_to_array, load_img
import glob
import matplotlib.pyplot as plt
import numpy as np
```

```python
datagen = ImageDataGenerator(
    rotation_range=20,  # rotation range within which to rotate pictures
    width_shift_range=0.2,  # range of horizontal translation
    height_shift_range=0.2,  # range of vertical translation
    rescale = 1./255,  # 
    shear_range=0,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='reflect')

face_image_path = '/media/tom/Files/MyPrograms/Python27/tensorflow_CNN/data/new_face_images/train_originals/face/'
none_image_path = '/media/tom/Files/MyPrograms/Python27/tensorflow_CNN/data/new_face_images/train_originals/none/'

# maybe do some splitting here for validation??
face_image_names = glob.glob(face_image_path + '*.jpg')
none_image_names = glob.glob(none_image_path + '*.jpg')
```

```python
# transform face labeled data
x = []
for image_name in face_image_names:
    #print(image_name)
    img = load_img(image_name)
    img_array = img_to_array(img)
    x.append(img_array)
x = np.array(x)

num_output_face = 100
i = 0
for batch in datagen.flow(x, batch_size=100, shuffle=True,
                          save_to_dir='./data/new_face_images/train/face/',
                          save_prefix='train', 
                          save_format='jpeg'):
    i += 1
    print('image generator batch: ' + str(i))
    if i > num_output_face:
        break
        
# transform none labeled data
x = []
for image_name in none_image_names:
    img = load_img(image_name)
    img_array = img_to_array(img)
    x.append(img_array)
x = np.array(x)

num_output_none = 100
i = 0
for batch in datagen.flow(x, batch_size=100, shuffle=True,
                          save_to_dir='./data/new_face_images/train/none/',
                          save_prefix='train',
                          save_format='jpeg'):
    i += 1
    print('image generator batch: ' + str(i))
    if i > num_output_none:
        break
```
